```python
import os
import re


doc_dict = {}

for file_name in os.listdir(data_dir):
    if not file_name.endswith(".txt"):
        continue

    with open(os.path.join(data_dir, file_name), "r", encoding="utf-8", errors="ignore") as f:
        content = f.read()
        
        messages = re.split(r"(?:^|\n)document_id:\s*(\d+)", content)


        for i in range(1, len(messages) - 1, 2):
            doc_id = int(messages[i])
            text = messages[i + 1].strip()
            doc_dict[doc_id] = text

df_list["text"] = df_list["document_id"].map(doc_dict)


print(df_list.shape)
df_list.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(stop_words='english', max_df=0.5)
X_tfidf = vectorizer.fit_transform(df_list["text"])

print("TF-IDF matrix shape:", X_tfidf.shape)
```

```python
from sklearn.decomposition import TruncatedSVD

svd = TruncatedSVD(n_components=2, random_state=42)
X_svd = svd.fit_transform(X_tfidf)

print("SVD-reduced shape:", X_svd.shape)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.scatterplot(x=X_svd[:, 0], y=X_svd[:, 1], hue=df_list['newsgroup'], palette='tab20', legend=None, alpha=0.7)
plt.title("2D SVD Projection of TF-IDF Vectors")
plt.xlabel("SVD Component 1")
plt.ylabel("SVD Component 2")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, adjusted_rand_score

kmeans = KMeans(n_clusters=20, random_state=42)
cluster_labels = kmeans.fit_predict(X_svd)

sil_score = silhouette_score(X_svd, cluster_labels)
print("Silhouette Score:", sil_score)

ari = adjusted_rand_score(df_list['newsgroup'], cluster_labels)
print("Adjusted Rand Index:", ari)
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x=X_svd[:, 0], y=X_svd[:, 1], hue=cluster_labels, palette='tab20', legend=None, alpha=0.7)
plt.title("KMeans Clustering (20 Clusters) on SVD-Reduced TF-IDF Data")
plt.xlabel("SVD Component 1")
plt.ylabel("SVD Component 2")
plt.tight_layout()
plt.show()
```

- TF-IDF gives a sparse high-dimensional matrix, which is hard to visualize.

- SVD helps reduce that into a compact, meaningful form.

- The 2D plot shows how similar or different the topics are without using labels.
