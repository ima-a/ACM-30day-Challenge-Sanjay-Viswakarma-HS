```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split



X = df.drop(columns=['Class'])
y = df['Class']

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42, stratify=y)
```

```python
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

linear_svc = LinearSVC(random_state=42, max_iter=5000)
linear_svc.fit(X_train, y_train)
y_pred_lin = linear_svc.predict(X_test)

print("LinearSVC Accuracy:", accuracy_score(y_test, y_pred_lin))
print(classification_report(y_test, y_pred_lin))
print(confusion_matrix(y_test, y_pred_lin))
```

```python
#rbf
from sklearn.svm import SVC

X = df.drop("Class", axis=1)
y = df["Class"]


X_sample, _, y_sample, _ = train_test_split(X, y, train_size=10000, stratify=y, random_state=42)

X_train_rbf, X_test_rbf, y_train_rbf, y_test_rbf = train_test_split(X_sample, y_sample, test_size=0.2, stratify=y_sample, random_state=42)

rbf_model = SVC(kernel="rbf", C=1.0, gamma="scale")
rbf_model.fit(X_train_rbf, y_train_rbf)
rbf_preds = rbf_model.predict(X_test_rbf)
print("SVM (RBF) Accuracy:", accuracy_score(y_test_rbf, rbf_preds))
print("Classification Report:\n", classification_report(y_test_rbf, rbf_preds))
print("Confusion Matrix:\n", confusion_matrix(y_test_rbf, rbf_preds))
```

```python
#poly
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

poly_svc = SVC(kernel='poly', degree=3, C=1.0, random_state=42)
poly_svc.fit(X_train, y_train)

poly_preds = poly_svc.predict(X_test)


poly_accuracy = accuracy_score(y_test, poly_preds)
print("Polynomial SVM Accuracy:", poly_accuracy)
print("Classification Report:\n", classification_report(y_test, poly_preds, zero_division=0))
print("Confusion Matrix:\n", confusion_matrix(y_test, poly_preds))
```
